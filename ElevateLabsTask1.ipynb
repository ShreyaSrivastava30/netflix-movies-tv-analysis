```python
## Some of the rows of this column has extra spaces which is creating problem in converting this column to datetime so lets remove those extra spaces and then change datatype

data.date_added = data.date_added.str.strip()
data.date_added = pd.to_datetime(data.date_added, format= '%B %d, %Y')
```

```python
## Finding number of null values in each column

data.isnull().sum()
```

```python
## Lets handle these null values one by one
## Lets look into director column, it has 30% missing values but dropping it wont be correct at this time, lets replace NAN to Unknown
## Doing same for cast and country

data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')
data['country'] = data['country'].fillna('Unknown')
```

```python
## Lets replace nulls of date_added, rating and duration with modes as these are really less in number as compared to previous columns with missing values

data['date_added'] = data['date_added'].fillna(data['date_added'].mode()[0])
data['rating'] = data['rating'].fillna(data['rating'].mode()[0])
data['duration'] = data['duration'].fillna(data['duration'].mode()[0])
```

```python
## Lets check missing values now

data.isnull().sum()
```
